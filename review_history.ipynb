```python
import numpy as np
import matplotlib.pyplot as plt
```

### Ganti path ini sesuai file yang mau dilihat

```python
# HISTORY_PATH = "vit_tiny_history.npy"
# HISTORY_NAME = "vit_tiny"
HISTORY_PATH = "deit_tiny_history.npy"
HISTORY_NAME = "deit_tiny"
```

```python
def main():
    history = np.load(HISTORY_PATH, allow_pickle=True).item()

    print("Key yang tersedia di history:")
    for k in history.keys():
        print(f"- {k} (len={len(history[k])})")

    train_loss = history.get("train_loss", None)
    val_loss   = history.get("val_loss", None)
    train_acc  = history.get("train_acc", None)
    val_acc    = history.get("val_acc", None)

    epochs = range(1, len(train_loss) + 1) if train_loss is not None else None

    if train_loss is not None and val_loss is not None:
        plt.figure()
        plt.plot(epochs, train_loss, label="Train Loss")
        plt.plot(epochs, val_loss, label="Val Loss")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.title("Train vs Val Loss" + f" ({HISTORY_NAME})")
        plt.legend()
        plt.tight_layout()
        plt.show()

    if train_acc is not None and val_acc is not None:
        plt.figure()
        plt.plot(epochs, train_acc, label="Train Acc")
        plt.plot(epochs, val_acc, label="Val Acc")
        plt.xlabel("Epoch")
        plt.ylabel("Accuracy")
        plt.title("Train vs Val Accuracy" + f" ({HISTORY_NAME})")
        plt.legend()
        plt.tight_layout()
        plt.show()

    print("\nSelesai menampilkan history.")

if __name__ == "__main__":
    main()
```
